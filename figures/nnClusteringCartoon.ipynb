```python
import matplotlib, math
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

plt.style.use('default')
font = {'size':14}
matplotlib.rc('font', **font)
matplotlib.rcParams['lines.linewidth'] = 4
```

```python

fig, ax = plt.subplots()
mu = 0
variance = .005
sigma = math.sqrt(variance)
x = np.linspace(0, 1, 100)
smDist = 0.15*stats.norm.pdf(x, mu, sigma)
signalDist1 = 0.1*stats.norm.pdf(x, 0.85, .6*sigma)
signalDist2 = 0.1*stats.norm.pdf(x, 0.8, .6*sigma)
plt.plot(x, smDist, label='SM background')
plt.plot(x, signalDist1, label='BSM model 1')
plt.plot(x, signalDist2, label='BSM model 2')
ax.set_xlim(0,1)
ax.set_ylim(0,1.2)
ax.set_xlabel('NN output')
ax.set_ylabel('Arbritrary units')
ax.legend(loc=0)
plt.savefig(f'nnCluster_cluster.pdf', bbox_inches='tight')
plt.savefig(f'nnCluster_cluster.svg', bbox_inches='tight')
```
